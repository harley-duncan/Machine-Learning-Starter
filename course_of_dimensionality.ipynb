```
import random
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from mpl_toolkits import mplot3d
```

Let's define the total number of points we want to display

```
POINTS = 50
```

# One dimension
Now, we want to generate random points. Since we are focusing on a single dimension, the value of `y` for every point will always be zero.

```
x = random.sample(range(0, 100), POINTS) 
y = np.zeros(POINTS)
```

Let's display our points in a single dimension.

Notice how a lot of the points are very close together. A clustering algorithm should be able to pick up some of these groups!

```
fig = plt.figure()
axis = plt.axes()
fig.set_size_inches(8, 1)

plt.scatter(x, y)

axis.set_xlim((0, 100))

plt.xticks([], [])
plt.yticks([], [])
plt.title("One dimension")
plt.show()
```

# Two dimensions

Let's now generate the same amount of points, but adding a second dimension to our data.

```
x = random.sample(range(0, 100), POINTS) 
y = random.sample(range(0, 100), POINTS) 
```

When we plot our points, we now see how the data is much more scattered. Remember we still have the same amount of points, but the second dimension has made those clusters much more harder to find.

```
fig = plt.figure()
axis = plt.axes()
fig.set_size_inches(8, 8)

plt.scatter(x, y)

axis.set_xlim((0, 100))
axis.set_ylim((0, 100))
axis.set_xlabel("Dimension 1", fontsize=12)
axis.set_ylabel("Dimension 2", fontsize=12)

plt.xticks([], [])
plt.yticks([], [])
plt.title("Two dimensions")
plt.show()
```

# Three dimensions

Let's add a third dimension.

```
x = random.sample(range(0, 100), POINTS) 
y = random.sample(range(0, 100), POINTS) 
z = random.sample(range(0, 100), POINTS)
```

When displaying our points, we can see that they are really sparse in the area of our chart.

```
fig = plt.figure()
axis = fig.add_subplot(1,1,1,projection='3d')

fig.set_size_inches(8, 8)

axis.scatter(x, y, z)

axis.set_xlim(0, 100)
axis.set_ylim(0, 100)
axis.set_zlim(0, 100)

axis.set_xlabel("Dimension 1", fontsize=12)
axis.set_ylabel("Dimension 2", fontsize=12)
axis.set_zlabel("Dimension 3", fontsize=12)

plt.title("Three dimensions")
plt.show()
```
